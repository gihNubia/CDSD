```python
from control import tf
import matplotlib.pyplot as plt
from control import step_response
from scipy import signal
import numpy as np
```

Função de transferencia: 
$$
H(s) = \frac{K_i}{J s^2 + C s + K}
$$
​
J = 0.01

C = 0.004

K = 10

Ki = 0.05


```python
from control import tf, step_response
import matplotlib.pyplot as plt

# Parâmetros
J = 0.01
C = 0.04
K = 10
Ki = 10

# Definir a função de transferência H(s) = Ki / (J * s^2 + C * s + K)
num = [Ki]
den = [J, C, K]
H = tf(num, den)

# Obter a resposta ao degrau
time, response = step_response(H)

# Criar o vetor da entrada do degrau (valor constante 1 a partir de t=0)
step_input = [1] * len(time)

# Plotar resposta ao degrau e entrada
plt.figure(figsize=(8, 6))
plt.plot(time, response, label="Resposta do Sistema", linewidth=2)
plt.plot(time, step_input, label="Entrada do Degrau", linestyle="--", color="red", linewidth=2)

# Configurar título, eixos e legenda
plt.title("Resposta ao Degrau e Entrada", fontsize=14)
plt.xlabel("Tempo (s)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
from control import tf, step_response, forced_response
import matplotlib.pyplot as plt

# Parâmetros
J = 0.01
C = 0.04
K = 10
Ki = 10

# Definir a função de transferência H(s) = Ki / (J * s^2 + C * s + K)
num = [Ki]
den = [J, C, K]
H = tf(num, den)

# Criar o vetor de tempo e a entrada senoidal (sin(t) com frequência omega)
tempo = np.linspace(0, 10, 1000)
omega = 1  # Frequência angular (rad/s)
entrada_senoidal = np.sin(omega * tempo)
#entrada_senoidal = np.exp(omega * tempo)

# Obter a resposta do sistema a essa entrada senoidal
tempo, resposta = forced_response(H, T=tempo, U=entrada_senoidal)

# Plotar resposta ao degrau e entrada
plt.figure(figsize=(8, 6))
plt.plot(tempo, resposta, label="Resposta do Sistema", linewidth=2)
plt.plot(tempo, entrada_senoidal, label="Entrada do Seno", linestyle="--", color="red", linewidth=2)

# Configurar título, eixos e legenda
plt.title("Resposta ao Seno e Entrada", fontsize=14)
plt.xlabel("Tempo (s)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

```python
from control import tf, step_response, forced_response
import matplotlib.pyplot as plt

# Parâmetros
J = 0.01
C = 0.04
K = 10
Ki = 10


# Definir a função de transferência H(s) = Ki / (J * s^2 + C * s + K)
num = [Ki]
den = [J, C, K]
H = tf(num, den)

# Criar o vetor de tempo e a entrada senoidal (sin(t) com frequência omega)
tempo = np.linspace(0, 10, 1000)
omega = 1  # Frequência angular (rad/s)
#entrada_senoidal = np.add(np.sin(omega * (tempo-3)), 1)
entrada_senoidal = [1 + (np.sin(omega*(T -1)) if T > 1 else 0) for T in tempo]

# Obter a resposta do sistema a essa entrada senoidal
tempo, resposta = forced_response(H, T=tempo, U=entrada_senoidal)

# Plotar resposta ao degrau e entrada
plt.figure(figsize=(10, 6))
plt.plot(tempo, resposta, label="Resposta do Sistema", linewidth=2)
plt.plot(tempo, entrada_senoidal, label="Entrada do Degrau", linestyle="--", color="red", linewidth=2)

# Configurar título, eixos e legenda
plt.xticks(np.arange(0, max(tempo)+1, 1))
plt.title("Resposta ao Degrau e Entrada", fontsize=14)
plt.xlabel("Tempo (s)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
